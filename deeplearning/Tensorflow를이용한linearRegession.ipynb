```python
# 딥러닝을 이해하기위한 척도로 이용
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
 # 가상적인데이터 생성
X=data= np.linspace(1,2,200)  # 시작 =1, 종료=2, 개수 = 200
y= X*4+np.random.randn(200)*0.3 # X를 4배, 편차 0.3정도
# #가우시안잡음추가
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(1, input_dim=1, activation = 'linear'))
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
model.fit(X,y, batch_size=1, epochs=30)
predict=model.predict(data)
# predict
plt.plot(X, predict, 'b', data, y, 'k.') #첫번째 그래프는 파란마커로, 두번째 그래프는 검정색으로 그리기
plt.show()
```
