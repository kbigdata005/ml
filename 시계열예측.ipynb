```python
import numpy as np 
import pandas as pd 

dates= pd.date_range('2020-01-01',periods=48,freq='M')

timestamp = np.arange(len(dates))
trend_factor = timestamp*1.1
cycle_factor = 10*np.sin(np.linspace(0,3.14*2,48))
season_factor = 7*np.sin(np.linspace(0,3.14*8,48))
np.random.seed(2004)
irrgular_factor = 2*np.random.randn(len(dates))

df = pd.DataFrame({'timeseries':trend_factor+cycle_factor+season_factor+irrgular_factor,
                   'trend':trend_factor,
                   'cycle':cycle_factor,
                   'seasonal':season_factor,
                   'irregular':irrgular_factor},
                   index=dates)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10 ,6))
df.timeseries.plot()
plt.title('timeseries')
plt.ylim(-12,55)
plt.show()
```
