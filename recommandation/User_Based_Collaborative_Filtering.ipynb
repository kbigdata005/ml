```python
# sparse matrix -> csr_matrix 이용
from scipy.sparse import csr_matrix

user_movie_rating_mat = csr_matrix(user_movie_rating_df.values)
# 방법 1 : cosine_similarity 
from sklearn.metrics.pairwise import cosine_similarity

cos_sim = cosine_similarity(user_movie_rating_mat)
movieid_series = pd.Series(movies_df['title'] , index=movies_df.index)

def all_movie_recommand(userId , sim_mat = cos_sim , sim_user=5):
    # 유사도가 가장큰 순서데로 5명의 유저 추출
    result = list(enumerate(cos_sim[userId]))
    result = sorted(result, key=lambda x: x[1] , reverse=True)
    user_sim_result = [x[0] for x in result[1:sim_user+1]]
    # print(user_sim_result)
    user_movie = set([x[0] for x in enumerate(list(user_movie_rating_df.iloc[userId , :])) if x[1] !=0 ])
    # 유사 유저가 본영화
    other_user_movie = []
    for i in user_sim_result:
        result = [x[0] for x in enumerate(list(user_movie_rating_df.iloc[i , :])) if x[1] !=0 ]
        other_user_movie += result
    other_movie = set(list(set(other_user_movie)))
    result_movie_index = list(other_movie - user_movie)
    # print(result_movie_index)
    return movieid_series[result_movie_index]
```

##### 가중평점을 이용하여 상위 10개 영화 추천


```python
# ratings_count , average_ratings
average_series = ratings_df.groupby('movieId')['rating'].mean()

count_series = ratings_df.groupby('movieId')['rating'].count()
df_1 = average_series.reset_index()
df_2 = count_series.reset_index()
df_3 = df_1.rename(columns={'rating':'average_rating'})
df_4 = df_2.rename(columns={'rating':'rating_count'})
df = pd.merge(df_3, df_4 , on='movieId')
merge_df = pd.merge(movies_df, df , on='movieId')
```

```python
merge_df.head(3)
```

```python
# 가중평점 계산해서 weighted_rating 컬럼 추가
C = merge_df['average_rating'].mean()
# 투표횟수 중 60%이상의 횟수에 달하는 숫자
# 예를들어 총 투표횟수가 100과 1일때 m값은 매우 달라진다.
m = merge_df['rating_count'].quantile(0.6)

def weighted_vote_average(record):
    v = record['rating_count']
    R = record['average_rating']
    
    return ( (v/(v+m)) * R) + ( (m/(m+v)) * C)
    
merge_df['weighted_rating'] = merge_df.apply(weighted_vote_average, axis=1)
```

###### 타겟 유저가 안본 영화를 비슷한 유저의 영화에 추출

```python
#타겟 유저가 본영화
user_movie = [x[0] for x in enumerate(list(user_movie_rating_df.iloc[0 , :])) if x[1] !=0 ]
# 유사 유저가 본영화
other_user_movie = []
for i in user_sim_result:
    result = [x[0] for x in enumerate(list(user_movie_rating_df.iloc[i , :])) if x[1] !=0 ]
    other_user_movie += result
# 타겟유저가 본영화를 제외한 영화를 유사유저에서 추출
result_movie_index = set(list(set(other_user_movie))) - set(user_movie)
result_movie_index_list = list(result_movie_index)
```

```python
def clean_data(x):
    return str.lower(x.replace("|" , " "))
reset_df['genres'] = reset_df['genres'].apply(clean_data)   
```

```python
reset_df[['average_rating' ,'rating_count' ]] = reset_df[['average_rating' ,'rating_count' ]].astype(str)
```
