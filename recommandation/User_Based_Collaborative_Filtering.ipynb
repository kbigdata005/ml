```python
# movie data import and index : user , column : movie , value : rating
import pandas as pd
import numpy as np

movies_df = pd.read_csv('../datasets/movies_small.csv')
ratings_df = pd.read_csv('../datasets/ratings_small.csv')
```

```python
movies_df.head(3)
```

```python
movies_df.shape
```

```python
ratings_df.head(3)
```

```python
ratings_df.drop('timestamp' , axis=1 , inplace=True)
```

```python
ratings_df.shape
```

```python
# ratings_df and movies_df merge 
movie_rating_df = pd.merge(movies_df , ratings_df , on='movieId')
```

```python
movie_rating_df.head(3)
```

```python
# pivot_table을 활용한  index : user , column : movie , value : rating DataFrame
user_movie_rating_df = movie_rating_df.pivot_table(values='rating' , index='userId' , columns='title')
```

```python
user_movie_rating_df.head(3)
```

```python
user_movie_rating_df.fillna(0 , inplace=True)
```

```python
user_movie_rating_df.head(3)
```

```python
# sparse matrix -> csr_matrix 이용
from scipy.sparse import csr_matrix

user_movie_rating_mat = csr_matrix(user_movie_rating_df.values)
```

```python
# 방법 1 : cosine_similarity 
from sklearn.metrics.pairwise import cosine_similarity

cos_sim = cosine_similarity(user_movie_rating_mat)
```

```python
len(cos_sim[0])
```

```python
enumerate(cos_sim[0])
```

```python
# 유사도가 가장큰 순서데로 5명의 유저 추출
result = list(enumerate(cos_sim[0]))
result = sorted(result, key=lambda x: x[1] , reverse=True)
user_sim_result = [x[0] for x in result[1:6]]
user_sim_result
```

##### 타겟 유저가 보지않은 영화를 비슷한유저가 본 영화 모두 추천

```python
#타겟 유저가 본영화
user_movie = [x[0] for x in enumerate(list(user_movie_rating_df.iloc[0 , :])) if x[1] !=0 ]
user_movie
```

```python
# 유사 유저가 본영화
other_user_movie = []
for i in user_sim_result:
    result = [x[0] for x in enumerate(list(user_movie_rating_df.iloc[i , :])) if x[1] !=0 ]
    other_user_movie += result
list(set(other_user_movie))
```
