```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies = pd.read_csv('../datasets/tmdb.csv')
movies.head()
```

```python
movies_5000 = pd.read_csv('../datasets/tmdb_5000_movies.csv')
movies_5000.head()
```

```python
import ast
def value_import(x):
    return [i['name'] for i in ast.literal_eval(x)]
movies_5000['genres'] = movies_5000['genres'].apply(value_import)
movies_5000['keywords'] = movies_5000['keywords'].apply(value_import)
```

```python
movies_5000.head(3)
```

```python
def join_words(x):
    return (' ').join(x)

movies_5000['genres_literal'] = movies_5000['genres'].apply(join_words)
```

```python
movies_5000['genres_literal']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# CountVectorizer를 적용하기 위해 공백문자로 word 단위가 구분되는 문자열로 변환
movies_5000['genres_literal'] = movies_5000['genres'].apply(lambda x : (' ').join(x))
count_vect = CountVectorizer(min_df=0.0, ngram_range = (1,2))
genre_mat = count_vect.fit_transform(movies_5000['genres_literal'])
print(genre_mat.shape)
```

```python
genre_sim = cosine_similarity(genre_mat ,genre_mat )
genre_sim
```

```python
# argsort() -> 유사도 높은거부터 정렬
genre_sim_sorted_ind = genre_sim.argsort()[:, ::-1]
print(genre_sim_sorted_ind[:1])
```

```python
# class sklearn.neighbors.NearestNeighbors(*, n_neighbors=5, radius=1.0, algorithm='auto', leaf_size=30, metric='minkowski', p=2, metric_params=None, n_jobs=None)
from sklearn.neighbors import NearestNeighbors

model_knn = NearestNeighbors(n_neighbors=20 , algorithm='auto', metric='cosine' , n_jobs=-1)
model_knn.fit(genre_mat)
```

```python
distances , indices_result = model_knn.kneighbors(genre_mat[0], n_neighbors=11)
```

```python
distances
```

```python
indices_result
```

```python
indices = pd.Series(movies_5000.index , index=movies_5000['original_title'])
```

```python
indices
```

```python
idx = indices['John Carter']
result = list(enumerate(genre_sim[idx]))
result = sorted(result, key=lambda x: x[1] , reverse=True)
# # print(result)
result = result[1:21]
print(result)
movie_indicies = [i[0] for i in result]
movie_indicies
```

```python
def get_recommandations(title , cosine_sim = genre_sim):
    idx = indices[title]
    result = list(enumerate(cosine_sim[idx]))
    result = sorted(result, key=lambda x: x[1] , reverse=True)
    # # print(result)
    result = result[1:21]
    # print(result)
    movie_indicies = [i[0] for i in result]
    movie_indicies
    return movies_5000['original_title'].iloc[movie_indicies]
```

```python
result = get_recommandations('John Carter')
print(result)
```

```python
C = movies_5000['vote_average'].mean()
# 투표횟수 중 60%이상의 횟수에 달하는 숫자
# 예를들어 총 투표횟수가 100과 1일때 m값은 매우 달라진다.
m = movies_5000['vote_count'].quantile(0.6)
def weighted_vote_average(record):
    v = record['vote_count']
    R = record['vote_average']
    
    return ( (v/(v+m)) * R) + ( (m/(m+v)) * C)
    
movies_5000['weighted_vote'] = movies_5000.apply(weighted_vote_average, axis=1)
```

```python
movies_5000.columns
```

```python
movies_5000.head(3)
```

```python
def get_recommandations_weighted_vote(title , cosine_sim = genre_sim):
    # title = str.lower(title.replace(" " , ""))
    idx = indices[title]
    result = list(enumerate(cosine_sim[idx]))
    result = sorted(result, key=lambda x: x[1] , reverse=True)
    # print(result)
    result = result[1:21]
    # print(result)
    movies_indicies = [i[0] for i in result]
    # print(book_indicies) df.iloc[similar_idx].sort_values(by=['weighted_vote'], ascending=False)[:top_n]
    # print(type(books.iloc[book_indicies]))
    # return books['original_title'].iloc[book_indicies]
    return movies_5000.iloc[movies_indicies].sort_values(by=['weighted_vote'], ascending=False)[:10]['original_title']
```

```python
result = get_recommandations_weighted_vote('John Carter')
print(result)
```

#### movies['soup']을 활용한 콘텐츠 기반 추천 시스템

```python
movies = pd.read_csv('../datasets/tmdb.csv' , index_col=0)
movies.head()
```

```python
def join_words(x):
    return ('').join(x)

movies['soup'] = movies['soup'].apply(join_words)
```

```python
movies['soup']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# CountVectorizer를 적용하기 위해 공백문자로 word 단위가 구분되는 문자열로 변환
count_vect = CountVectorizer(min_df=0.0, ngram_range = (1,2))
soup_mat= count_vect.fit_transform(movies['soup'])
print(soup_mat.shape)
soup_mat
```

```python
# class sklearn.neighbors.NearestNeighbors(*, n_neighbors=5, radius=1.0, algorithm='auto', leaf_size=30, metric='minkowski', p=2, metric_params=None, n_jobs=None)
from sklearn.neighbors import NearestNeighbors

model_knn = NearestNeighbors(n_neighbors=20 , algorithm='auto', metric='cosine' , n_jobs=-1)
model_knn.fit(soup_mat)
```

```python
distances , indices_result = model_knn.kneighbors(soup_mat[1], n_neighbors=11)
```

```python
indices_result
```

```python
distances
```

```python
indices = pd.Series(movies.index , index=movies['original_title'])
idx = indices['John Carter']
distances , indices_result = model_knn.kneighbors(soup_mat[0], n_neighbors=11)
result = indices_result[0][1:]
similiarity = list(1-np.array(distances))[0]
similiarity
```

```python

def get_recommandations(title , model=model_knn):
    idx = indices[title]
    distances , indices_result = model_knn.kneighbors(soup_mat[idx], n_neighbors=11)
    result = indices_result[0][1:]
    similiarity = list(1-np.array(distances))[0]
    return (movies['original_title'].iloc[result] , similiarity)
```

```python
title_series , sim = get_recommandations('John Carter')
# list(title_series)
for i in range(len(title_series)):
    print(f'{i+1}번째 추천영화는 \n')
    print(f'영화 {list(title_series)[i]}이고 유사도는 {sim[i]}')
    print(f'=================== \n')
```
